```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

# **Loading The Dataset**
# Loading a CSV file into a pandas DataFrame

```python
df = pd.read_csv("/content/drive/MyDrive/ML Datasets/Heart Disease dataset.csv")
```

# Spliting and scaling the data

```python
# Here the data will be split into input features and output features.
# Here we drop target from X as X represents input factors.
# Similarly we define Y as target as Y represents the output features.
X = df.drop('target', axis=1)
Y = df['target']
```

# Initializing a Standard Scaler object

```python
# Here, we have scaled the input features according to their contribution to the final output.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Generating a heatmap to visualize the correlation between the features


```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True)
plt.title("Feature Correlation with Target")
plt.show()
```

# **Training the Model**

# Splitting the dataset into training and testing sets

```python
# Here we have used random state = 42 to reproduce the same results everytime we run the project.
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

# Training and evaluating Support Vector Classifier model

```python
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(Xtrain, Ytrain)
```

# Evaluating the model to print a classification report

```python
Ypred = svm_model.predict(Xtest)
print("Accuracy:", accuracy_score(Ytest, Ypred))
print("\nClassification Report:\n", classification_report(Ytest, Ypred))
```

# Displaying a confusion matrix

```python
cm = confusion_matrix(Ytest, Ypred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
# Here fmt = 'd' helps to convert the output data to integers.

plt.title("SVM Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# Trying with different Kernels using Support Vector Classifier Model

```python
# Here we compare the accuracy of the model by fitting it using different kernels to get the best results.
# Here we created two lists of kernels and accuracies which will be appended as we fit the model with each kernel.

kernels = []
accuracies = []

for kernel in ['linear', 'poly', 'rbf', 'sigmoid']:
    model = SVC(kernel=kernel)
    model.fit(Xtrain, Ytrain)
    pred = model.predict(Xtest)
    acc = accuracy_score(Ytest, pred)
    kernels.append(kernel)
    accuracies.append(acc)
    print(f"Kernel: {kernel} - Accuracy: {acc}")
```

# Generating Bar Plot to compare the accuracy of the four Kernels

```python
plt.figure(figsize=(8, 6))
plt.bar(kernels, accuracies, color=['pink', 'lightgreen', 'purple', 'salmon'])
plt.title("SVM Model Accuracy by Kernel Type")
plt.xlabel("Kernel Type")
plt.ylabel("Accuracy")
plt.ylim(0.7, 1)
plt.show()
```

# Calculating and visualizing the feature importances from the trained Support Vector Classifier model

```python
# Here we have plotted a bar graph comparing the contribution of different features towards the final output.
svm_linear = SVC(kernel='linear', random_state=42)
svm_linear.fit(Xtrain, Ytrain)

importances = np.abs(svm_linear.coef_[0])
# Here coef refers to magnitude of coefficient which represents the importance of each feature in a lineal model.

features = X.columns
indices = np.argsort(importances)[::-1]
# Here we sort the features importance in the descending order and return the indices.

plt.figure(figsize=(10, 6))
plt.title("Feature Importances - Linear SVM")
plt.bar(range(X.shape[1]), importances[indices])
plt.xticks(range(X.shape[1]), features[indices], rotation=90)
# Here we rotate the labels by 90 degrees to make them readable.

plt.xlabel("Features")
plt.ylabel("Importance")
plt.tight_layout()
# Here we adjust the plot parameters automatically.

plt.show()
```

# Calculating the permutation importances of SVM RBF Kernel

```python
from sklearn.inspection import permutation_importance

r = permutation_importance(svm_model, Xtest, Ytest, n_repeats=10, random_state=42)
# Here we calculate permutation importance for the RBF SVM model.

feature_importances = pd.Series(r.importances_mean, index=X.columns)
# Here we store feature names and their importances.

display(feature_importances.sort_values(ascending=False))
# Here we sort importances and display.
```

# Visualizing the permutation importances of SVM RBF Kernel

```python
plt.figure(figsize=(10, 6))
feature_importances.sort_values(ascending=False).plot.bar()
plt.title("Feature Importances - RBF SVM (Permutation Importance)")
plt.xlabel("Features")
plt.ylabel("Importance")
plt.tight_layout()
# Here we adjust the plot parameters automatically.

plt.show()
```

# **Predicting on New Data**

```python
new_patient = pd.DataFrame([{
    'age': 65,
    'sex': 1,
    'chest pain type': 4,
    'resting bp s': 125,
    'cholesterol': 0,
    'fasting blood sugar': 1,
    'resting ecg': 0,
    'max heart rate': 176,
    'exercise angina': 0,
    'oldpeak': 1.6,
    'ST slope': 1}])
# Here we enter the new patient data.

new_patient_scaled = scaler.transform(new_patient)
# Here we scale the input data as model was trained on scaled data.

prediction = svm_model.predict(new_patient_scaled)
# Here the model is trained to make the prediction.

new_patient['prediction'] = prediction
new_patient['prediction'] = new_patient['prediction'].map({0: "No Heart Disease", 1: "Heart Disease"})
# Here we map the output to the human readable string.

print("Patient Data with Prediction:")
display(new_patient)
```
